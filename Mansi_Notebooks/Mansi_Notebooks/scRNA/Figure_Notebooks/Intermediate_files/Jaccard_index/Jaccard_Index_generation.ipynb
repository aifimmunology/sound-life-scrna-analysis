# Jaccard function

```r
# #Function to calculate Jaccard index
# jaccard_index <- function(set1, set2) {
#   intersection <- intersect(set1, set2)
#   union <- union(set1, set2)
#   return(length(intersection) / length(union))
# }
```

```r
jaccard <- function(a, b) {
    intersection = length(intersect(a, b))
    union = length(a) + length(b) - intersection
    return (intersection/union)
}
```

```r
# library(sets)
# library(dplyr)

# Initialize an empty dataframe to store results
results <- data.frame(celltype = character(),
                      jaccard_similarity = numeric(),
                       total_DEGs_male = integer(),
                       total_DEGs_female = integer(),
                       common_DEGs = integer(),
                       stringsAsFactors = FALSE)

# Iterate through cell types
for (cell in unique(master$celltype)) {
  
    # Filter data for the current cell type
  cell_filt <- master %>% dplyr::filter(celltype == cell, padj <= 0.1)
  
  # Filter DEGs for males and females
  DEG_male <- cell_filt[cell_filt$padj <= 0.1 & cell_filt$type == "DEGs_Male", "X"]
  DEG_female <- cell_filt[cell_filt$padj <= 0.1 & cell_filt$type == "DEGs_Female", "X"]
  
  # Remove NA values from DEGs
  DEG_male <- na.omit(DEG_male)
  DEG_female <- na.omit(DEG_female)
  
  # Convert filtered DEGs to sets
  set_DEG_male <- as.set(DEG_male)
  set_DEG_female <- as.set(DEG_female)
  
  # Calculate total DEGs counts
  total_DEGs_male <- length(DEG_male)
  total_DEGs_female <- length(DEG_female)
  
  # Calculate common DEGs counts
  common_DEGs <- length(intersect(set_DEG_male, set_DEG_female))
  
  # Calculate Jaccard similarity
  jaccard_sim <- jaccard(set_DEG_male, set_DEG_female)
  
  # Append results to dataframe
  results <- rbind(results, data.frame(celltype = cell,
                                       jaccard_similarity = jaccard_sim,
                                        total_DEGs_male = total_DEGs_male,
                                        total_DEGs_female = total_DEGs_female,
                                        common_DEGs = common_DEGs))
}

# Print results
print(results)

```
