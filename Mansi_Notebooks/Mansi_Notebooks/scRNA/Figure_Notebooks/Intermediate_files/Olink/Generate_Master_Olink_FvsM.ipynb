```r
res_all_female<-olink_female %>% filter(factor=='Age')
res_all_female
```

```r
data <- res_all_female%>% mutate(Expression = case_when(diff >=0.1 & padj <= 0.05 ~ "Up-regulated-BR2",
                                                     diff <= -0.1 & padj <= 0.05 ~ "Down-regulated-BR2",
                                                     TRUE ~ "Unchanged"))
```

```r
top <- 30
top_genes <- bind_rows(
  data %>% 
    filter(Expression == 'Up-regulated-BR2') %>% 
    arrange(padj, desc(abs(diff))) %>% 
    head(top),
  data %>% 
    filter(Expression == 'Down-regulated-BR2') %>% 
    arrange(padj, desc(abs(diff))) %>% 
    head(top))
```
