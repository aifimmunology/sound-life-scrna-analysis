### Clean up and create pseudobulked counts of average gene expression across samples
Gene list based on [this spreadsheet](https://docs.google.com/spreadsheets/d/1XTEy0aurnNpFAYdECb3JXqOBM6-uSUKGyGtlOgHq6cc/edit#gid=0).

Steps here: 
1. Identify genes of interest from DIHA.
2. Compute pseudobulked gene expression.
    - Formula: Mean expression of genes per sample (each donor) 
3. Add necessary metadata.

````python
# Get cell types and expression data
sample_id = 'sampleID' 
cell_type = 'celltypist_l3' 

## Genes of interest from spreadsheet
genes = ['G0S2', 
         'TNF', 
         'IL1B', 
         'NR4A1', 
         'PTGS2', 
         'BCL2A1', 
         'NR4A1', 
         'METRNL', 
         'FKBP5', 
         'RBP7', 
         'FCGR3A', 
         'CCR1', 
         'RASGEF1B', 
         'CXCL8', 
         'NLRP3', 
         'CCL4']
````

````python
samples = adata.obs[sample_id]
expr_data = adata.X

# Calculate average expression for each cell type
avg_expression = pd.DataFrame(0, index=adata.var_names, columns=np.unique(samples)) ## Dummy DF with sample IDs
for type in np.unique(samples):
    sample = np.where(samples == type)[0]
    avg_expression[type] = np.mean(expr_data[sample, :], axis=0)

# Transpose avg_expression and reset index
avg_expression = avg_expression.T.reset_index()

# Add metadata from obs
mapping_df = adata.obs[['sampleID', 'age_enrollment', 'biological_sex']].drop_duplicates().set_index('sampleID')
avg_expression['age_enrollment'] = avg_expression['index'].map(mapping_df['age_enrollment']).replace('89+', 89).astype(float)
avg_expression['biological_sex'] = avg_expression['index'].map(mapping_df['biological_sex'])

# Melt the DataFrame to have genes in one column
melted_df = avg_expression.melt(id_vars=['age_enrollment', 'biological_sex'], 
                                value_vars=genes, 
                                var_name='gene', 
                                value_name='expression')
````

### Sex based changes in GEX for IL1B+ CD14 Monocytes

### Add an unweighted composite score of mean gene expression (Based on chat with Samir)
Logic: By adding effects (such as changes in GEX), we should see a stronger variance if the effect is more pronouced in one cohort vs the other). A simple average is taken across a set of genes for each cohort, female and male. Here, we hope to see a bias towards females. 

> If gene A, B, C and D contribute to difference in group 1 vs group 2, then mean(A,B,C,D) between the 2 groups should have a notable difference. 

````python
samples = adata.obs[sample_id]
expr_data = adata.X

avg_expression = pd.DataFrame(0, index=adata.var_names, columns=np.unique(samples))
for type in np.unique(samples):
    sample = np.where(samples == type)[0]
    avg_expression[type] = np.mean(expr_data[sample, :], axis=0)

avg_expression = avg_expression.T.reset_index()

idx_genes = ['index']+genes
df_subset = avg_expression[idx_genes]

sample_means = df_subset.iloc[:, 1:].mean(axis=1)
df_subset['mean_gex'] = sample_means

mapping_df = adata.obs[['sampleID', 'age_enrollment', 'biological_sex']].drop_duplicates().set_index('sampleID')
df_subset['age_enrollment'] = df_subset['index'].map(mapping_df['age_enrollment']).replace('89+', 89).astype(float)
df_subset['biological_sex'] = df_subset['index'].map(mapping_df['biological_sex'])
````

#### Plot linear effects 

````python
genes = ['mean_gex']
melted_df = df_subset.melt(id_vars=['age_enrollment', 'biological_sex'], value_vars=genes, var_name='gene', value_name='expression')

g = sns.FacetGrid(melted_df, 
                  row='gene',
                  hue='biological_sex', 
                  height=4, 
                  aspect=1.6, 
                  palette={'Female': '#690447', 'Male': '#0e5e9c'}, 
                  sharey=False)

g.map(sns.regplot, 'age_enrollment', 'expression', scatter_kws={'s': 5})
g.add_legend()

plt.show()
````

#### Plot non-linear effects 

````python
def plot_loess(x, y, **kwargs):
    lowess = sm.nonparametric.lowess(y, x)
    plt.plot(lowess[:, 0], lowess[:, 1], color='red')

g = sns.FacetGrid(melted_df, 
                  row='gene',
                  hue='biological_sex', 
                  height=4, 
                  aspect=1.6, 
                  palette={'Female': '#690447', 'Male': '#0e5e9c'}, 
                  sharey=False)

g.map(sns.scatterplot, 'age_enrollment', 'expression')
g.map(plot_loess, 'age_enrollment', 'expression')
g.add_legend()

plt.show()
````
