## Helper functions

```r
cache_uuid_path <- function(uuid) {
    if(!dir.exists(paste0("cache/", uuid))) {
        hise_res <- cacheFiles(list(uuid))
    }
    
    cache_path <- paste0("cache/",uuid)
    cache_file <- list.files(cache_path, full.names = TRUE)
    
    cache_file
}
```

```r
stored_file_df <- function(store_name) {
    ps_files <- listFilesInProjectStores(list(store_name))
    ps_files <- map(
        ps_files$files, 
        function(l) {
            l <- l[c("id", "name")]
            as.data.frame(l)
        }) %>%
      list_rbind()
    
    ps_files
}
```

```r
library(dplyr)
library(tidyr)

top_degs <- sex_degs %>%
  dplyr::group_by(celltype) %>%
  dplyr::arrange(padj) %>%
  dplyr::slice_head(n = 5) %>%
  ungroup()

# Create a list of top 10 DEGs for each cell type
degs_list <- top_degs %>%
  dplyr::group_by(celltype) %>%
  dplyr::summarize(genes = list(gene)) %>%
  pull(genes)

# Identify shared DEGs across all cell types
shared_degs <- Reduce(intersect, degs_list)

# Calculate the number of shared DEGs for each cell type
shared_degs_count <- top_degs %>%
  dplyr::filter(gene %in% shared_degs) %>%
  dplyr::group_by(celltype) %>%
  dplyr::summarize(shared_degs = n())

# Print the result
print(shared_degs_count)
```

# Heatmap

```r
top_degs <- sex_degs %>%
  dplyr::group_by(celltype) %>%
  #filter(padj <0.05) %>%
  dplyr::arrange(padj) %>%  #rows with the smallest padj values will appear first
  dplyr::slice_head(n = 5) %>%
  ungroup()
```

```r
unique(top_degs$gene)
length(unique(top_degs$gene))
```

```r
unique(top_degs$gene)
length(unique(top_degs$gene))
```

```r
library(tidyr)
library(pheatmap)

# Create a binary matrix of shared DEGs
shared_degs_matrix <- top_degs %>%
  dplyr::filter(gene %in% top_degs$gene) %>%
  dplyr::mutate(present = 1) %>%
  pivot_wider(names_from = gene, values_from = present, values_fill = 0)

# # Remove celltype column and convert to matrix
# row_names <- shared_degs_matrix$celltype
# shared_degs_matrix <- shared_degs_matrix %>% select(-celltype) %>% as.matrix()
# rownames(shared_degs_matrix) <- row_names

# # Plot the heatmap
# pheatmap(shared_degs_matrix, cluster_rows = TRUE, cluster_cols = TRUE, 
#          show_rownames = TRUE, show_colnames = FALSE,
#          main = "Shared Top 10 DEGs in F vs M across Cell Types")
```

```r
head(shared_degs_matrix)
```

```r
# Columns to remove
cols_to_remove <- c("X", "baseMean", "log2FoldChange", "lfcSE", "stat", "pvalue", "padj", "contrast", "Direction")

# Remove columns
shared_degs_matrix_v2 <- shared_degs_matrix[ , setdiff(names(shared_degs_matrix), cols_to_remove)]
```

```r
shared_degs_matrix_v2
```

```r
# Assuming your data frame is named `df`

# Group by celltype and summarise counts for each gene
consolidated_df <- shared_degs_matrix_v2 %>%
  dplyr::group_by(celltype) %>%
  dplyr::summarise_all(sum)

# Reset row names
rownames(consolidated_df) <- NULL

# View the consolidated dataframe
print(consolidated_df)
dim(consolidated_df)
```

```r
# Remove celltype column and convert to matrix
row_names <- consolidated_df$celltype
consolidated_df <- consolidated_df %>% select(-celltype) %>% as.matrix()
rownames(consolidated_df) <- row_names
```

```r
head(consolidated_df)
```

```r
# generate heatmap
library(RColorBrewer)
library(viridis)
```

```r
# Plot the heatmap
breaks_sequence <- seq(2, -2, length.out = 100)
options(repr.plot.width = 30, repr.plot.height = 5)
pheatmap(shared_degs_matrix_v2, cluster_rows = TRUE, cluster_cols = TRUE, 
         show_rownames = TRUE, show_colnames = FALSE,
         main = "Shared Top 10 DEGs in F vs M across Cell Types")
```
