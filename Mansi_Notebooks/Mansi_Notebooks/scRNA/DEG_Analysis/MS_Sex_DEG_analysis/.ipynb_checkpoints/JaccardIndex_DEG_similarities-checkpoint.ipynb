## Jaccard function

```r
#Function to calculate Jaccard index
jaccard_index <- function(set1, set2) {
  intersection <- intersect(set1, set2)
  union <- union(set1, set2)
  return(length(intersection) / length(union))
}
```

```r
jaccard <- function(a, b) {
    intersection = length(intersect(a, b))
    union = length(a) + length(b) - intersection
    return (intersection/union)
}
```

```r
list_to_set <- function(x) {
  unique(unlist(x))
}
```

```r
# Function to compare DEG files in Male and Female folders
compare_DEG_files <- function(file_name, male_folder, female_folder) {
  male_file <- file.path(male_folder, file_name)
  female_file <- file.path(female_folder, file_name)
  
  if (!file.exists(male_file) | !file.exists(female_file)) {
    return(NULL)
  }
  
  male_DEG <- read.csv(male_file)
  female_DEG <- read.csv(female_file)
  
  jaccard_index <- calculate_jaccard_index(male_DEG, female_DEG)
  total_DEGs_male <- sum(male_DEG$padj <= 0.1)
  total_DEGs_female <- sum(female_DEG$padj <= 0.1)
  
  return(list(jaccard_index = jaccard_index, total_DEGs_male = total_DEGs_male, total_DEGs_female = total_DEGs_female))
}

# Create an empty data frame to store results
results_df <- data.frame(cell_type = character(), jaccard_index = numeric(), total_DEGs_male = integer(), total_DEGs_female = integer(), stringsAsFactors = FALSE)

# Compare DEG files for each cell type
for (file_name in male_files) {
  cell_type <- str_replace(file_name, "_Y1D0\\.csv$", "")  # Extract cell type from file name
  comparison_result <- compare_DEG_files(file_name, male_folder, female_folder)
  if (!is.null(comparison_result)) {
    results_df <- rbind(results_df, data.frame(cell_type = cell_type, 
                                               jaccard_index = comparison_result$jaccard_index,
                                               total_DEGs_male = comparison_result$total_DEGs_male,
                                               total_DEGs_female = comparison_result$total_DEGs_female))
  }
}

# Print or further process the results
print(results_df)

```
