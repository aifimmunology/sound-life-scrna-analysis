# Get Flow data

```r
meta_data<-read.csv("/home//jupyter/BRI_Claire/Flow_Locked/flow_meta_data.csv")
CMV_INFO<-read.csv ("/home//jupyter/BRI_Claire/scRNA_Locked/CMV_20220718.csv")
meta_data<-left_join(meta_data,CMV_INFO[2:3])
```

```r
seleced_kit_id<-pull(read.csv('/home//jupyter/BRI_Claire/scRNA_Locked/hise_meta_data_2023-08-09.csv')['sample.sampleKitGuid'])
meta_data<-meta_data %>% filter(sample.sampleKitGuid %in% seleced_kit_id)

seleced_kit_id<-as.data.frame(table(meta_data$file.batchID,meta_data$sample.sampleKitGuid)) %>% 
                filter(Freq>=4) %>% 
                select(Var2) %>% 
                pull()
meta_data<-meta_data %>% filter(sample.sampleKitGuid %in% seleced_kit_id)
```

```r
meta_data<-meta_data[!duplicated(meta_data[c('sample.sampleKitGuid','file.panel')]),]
```

```r
res_list<-mclapply(1:dim(meta_data)[1], function(i){

df<-read.csv(paste0('/home//jupyter/BRI_Claire/Flow_Locked/',meta_data$file.path[i]))
df$ panel<-meta_data$file.panel[ i]

return(as.data.frame(df))
    
},mc.cores=60)
```

```r
flow_freq<-do.call(rbind,res_list)
```

```r
flow_freq$sample.sampleKitGuid<-paste0('KT',substr(flow_freq$sample_id,3,7))
```

```r
flow_freq<-flow_freq %>% filter(sample.sampleKitGuid%in% unique(CBC$sample.sampleKitGuid))
```

# CBC counts of Monocytes

```r
flow_freq_Mono<-flow_freq %>% filter(panel=='PM1',labels%in% c('cd14_monocytes','cd16_monocytes','int_monocytes'))
```

```r
flow_freq_Mono <- reshape(flow_freq_Mono[c('labels','counts','sample.sampleKitGuid')], idvar="sample.sampleKitGuid", timevar="labels", v.names="counts", direction="wide", sep="_")
```

```r
total_counts <- rowSums(flow_freq_Mono[, -1])  
flow_freq_Mono[, -1] <- sweep(flow_freq_Mono[, -1], 1, total_counts, FUN = function(x, y) (x / y) )
```

```r
CBC<-left_join(CBC,flow_freq_Mono)
```

```r
CBC[, 7:9] <- round(CBC[,  7:9] * CBC$absolute_monocyte_count_amc,0)
```

# CBC counts of lymphocyte
