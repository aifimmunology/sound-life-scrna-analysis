```r
cv_all<-as.data.frame(read_parquet('cv.parquet'))
```

```r
meta_data_scRNA<-read.csv("hise_meta_data_2023-08-09.csv")
CMV=read.csv("CMV_20220718.csv") 
meta_data_scRNA<-left_join(meta_data_scRNA,CMV,by = c("subject.subjectGuid"))
```

```r
meta<-unique(meta_data_scRNA[c('subject.subjectGuid','CMV.IgG.Serology.Result.Interpretation','subject.biologicalSex','cohort.cohortGuid')])
```

```r
colnames(meta)<-c('donor','CMV','Sex','cohort.cohortGuid')
```

```r
cv_all<-left_join(cv_all,meta)
```

```r
options(repr.plot.width = 15, repr.plot.height = 25)
P<-ggplot(data=cv_all, aes(x=cv, group=cohort.cohortGuid, fill=cohort.cohortGuid)) +
    geom_density(adjust=1.5, alpha=.4) +
    theme_bw(base_size=14)+facet_wrap(~celltype,ncol=4, scales='free')
P
```

```r
options(repr.plot.width = 15, repr.plot.height = 5)
P<-ggplot(data=cv_all %>% filter(celltype %in% c('B naive','CD14 Mono','CD16 Mono'
                                                ,'CD4 TCM','CD4 TEM','NK')), aes(x=cv, group=cohort.cohortGuid, fill=cohort.cohortGuid)) +
    geom_density(adjust=1.5, alpha=.4) +
    theme_bw(base_size=14)+facet_wrap(~celltype,ncol=3, scales='free')
P
```

```r
options(repr.plot.width = 15, repr.plot.height = 5)
P<-ggplot(data=cv_all %>% filter(celltype %in% c('MAIT','gdT','CD8 Naive'
                                                ,'Treg Naive','Treg Memory','CD8 TEMRA')), aes(x=cv, group=cohort.cohortGuid, fill=cohort.cohortGuid)) +
    geom_density(adjust=1.5, alpha=.4) +
    theme_bw(base_size=14)+facet_wrap(~celltype,ncol=3, scales='free')
P
```

```r
options(repr.plot.width = 6, repr.plot.height = 5)
P<-ggplot(data=cv_all %>% filter(celltype %in% c('MAIT','gdT')), aes(x=cv, group=cohort.cohortGuid, fill=cohort.cohortGuid)) +
    geom_density(adjust=1.5, alpha=.4) +
    theme_bw(base_size=14)+facet_wrap(~celltype,ncol=1, scales='free')
P
```
