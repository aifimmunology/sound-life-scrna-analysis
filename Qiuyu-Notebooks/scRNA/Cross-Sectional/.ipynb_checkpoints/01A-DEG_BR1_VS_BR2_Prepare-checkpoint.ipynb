```python
adata=sc.read_h5ad('All_Raw_20230825.h5ad')
```

```python
cross_tab_prop = pd.crosstab(index=adata.obs['predicted.celltype.l2.5'],columns=adata.obs['sample.visitName'],normalize="index")
cross_tab_prop.plot(kind='bar', stacked=True, colormap='tab20',figsize=(8, 5))
plt.legend(bbox_to_anchor=(1, 1),ncol=1, loc="upper left")
plt.ylabel("Proportion")
plt.show()
```
